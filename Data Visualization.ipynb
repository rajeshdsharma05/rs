<h4> Load data and store in dataframe </h4>

```python
# Encode categorical variables
df['Ticker Symbol'] = df['Ticker Symbol'].astype('category')
df['Security'] = df['Security'].astype('category')
df['GICS Sector'] = df['GICS Sector'].astype('category')
df['GICS Sub Industry'] = df['GICS Sub Industry'].astype('category')
df.info()
```

### 2. Reason of chart selection :

### 3. Pre-attentive attributes used :  

1. Shape : Shape of the curve depicted brings quick attention to the type of distribution of data.  


2. Length : Length of the bars (bins) brings attention to the frequency or density of values within each range.Longer 
    bars correspond to a higher frequency or density, while shorter bars correspond to a lower frequency or density. 
   

### 4. Gestalt principles used :

1. Continuity : The bars are arranged in a continuous line or curve, with no gaps between them. This continuity helps 
to create the perception of a smooth distribution of values, and makes it easier to identify any peaks or valleys in the data.



2. Proximity: The bars are arranged side by side, with each bar representing a range of values. This proximity between bars helps to create the perception of discrete groups or bins, and emphasizes the differences in frequency or density between them.

### 3. Pre-attentive attributes used :  

1. Intensity : The bold datapoints in the chart brings attention to the outliers and their intensity .  


2. Enclosure : The box and whisker plot creates an enclosure for datapoints that are not outliers and presents clear demarcation.

1. Focal Point : Outliers are marked in black color to create focal point and distguish them as outliers

```python
avgCashRatio = df.groupby('GICS Sector', as_index=False)[['Cash Ratio']].mean()
sorted_avgCR = avgCashRatio.sort_values('Cash Ratio', ascending=True)
ax = sorted_avgCR.plot(kind='barh', x='GICS Sector', figsize=(8,6))
ax.set_title('\n Average Cash Ratio - Variation - Across Sectors \n ')
ax.grid(False)
ax.legend().remove()
ax.set_facecolor('white')
ax.set_ylabel('Economic Sector')
ax.set_xlabel('\n Average Cash Ratio \n')
ax.axvline(x=100, color='g', linestyle='--');
```

### Answer 4  


### 1. Summary of findings :  
        
   Average Cash Ratio varies significantly across sectors
        
   It is highest in Information Technology while it is lowest in Utilities.
        
   Energy, Real Estate and Consumer Dicretionary have almost equal avearage cash ratio.
   
        
### 2. Reason of chart selection : 

   Horizontal bar chart is selected as the values are numerical and descending order sorted chart gives a simple view of variation of avearge cash ratio across sectors. Sectors names are clearly readable depicted horizontally involving less cognitive load.
    

### 3. Pre-attentive attributes used :  

1. Length : The length of the bars is conveying quantitative information. Longer bars indicate higher values, and shorter bars indicate lower values.


2. Color : The green color threshold line at 100 indicates that companies in those sectors has more cash and cash equivalents than its current liabilities, which could indicate a healthy financial position. This threshold line is a useful pre-attentive attribute.
   

### 4. Gestalt principles used :

1. Focal Point : Threhold line with green color creates a focal point.

2. Pragnaz: Meaningful data presented in simple and consie manner.

```python
# Calculate percentage distribution of companies by sector
sector_counts = df['GICS Sector'].value_counts(normalize=True) * 100
sorted_sector_counts = sector_counts.sort_values()

# Create stacked bar chart
plt.figure(figsize=(10, 6))
sorted_sector_counts.plot(kind='barh', stacked=True)
plt.xlabel("% Distribution of Companies")
plt.ylabel("Sector \n")
plt.title("\n Percentage Distribution of Companies by Sector \n")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

### Answer 5  


### 1. Summary of findings :  
        
   Industrials and Financials amounts to 30% to total companies. 
   
        
### 2. Reason of chart selection : 

   Staked bar chart is selected as the percentage values in descending order sorted chart gives a simple view of percentage distribution of companies across sectors. Also, sectors names are clearly readable depicted horizontally involving less cognitive load.
    

### 3. Pre-attentive attributes used :  

1. Length : The length of the bars is conveying quantitative information. Longer bars indicate higher values, and shorter bars indicate lower values.


2. Lines : Grid lines are used to make viewer know the percentage value.

1. Continuity : Alignment of bars stacked together facilitate comparision of different sectors in terms of % distribution of companies.

### Question 6

### Group's choice EDA question :   

What is the distribution of Volatility across different GICS Sectors?

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x="Volatility", y="GICS Sector", data=df)
plt.title("\n Distribution of Volatility by GICS Sector \n", fontsize=16)
plt.xlabel("Volatility", fontsize=14)
plt.xticks(rotation=90)
plt.ylabel("GICS Sector", fontsize=14)

# Adjust the layout of the plot
plt.tight_layout()

# Display the plot
plt.show()
```

### Answer 6 


### 1. Summary of findings :  
        
   Based on the box plot, we can see that the Energy and Materials sectors have the highest volatility, while the Real Estate and Utilities sectors have the lowest volatility. The Consumer Staples sector also appears to have relatively low volatility. 
   
        
### 2. Reason of chart selection : 

   The reason for selecting a box plot is that it allows us to compare the distribution of a numerical variable across different categories. The box plot shows the median, quartiles, and any outliers for each category, which makes it easy to see the differences in distribution.
    

### 3. Pre-attentive attributes used :  

The pre-attentive attributes used in the box plot are position and length. The position of the boxes and whiskers on the y-axis shows the magnitude of the volatility for each sector, while the length of the boxes and whiskers shows the range of the data.
   

### 4. Gestalt principles used :

The gestalt principles used in the box plot are similarity and proximity. The boxes and whiskers for each sector are similar in shape and size, which makes it easy to compare them visually. The boxes and whiskers are also grouped closely together on the x-axis, which emphasizes their relatedness and makes it easy to see the differences between them.
